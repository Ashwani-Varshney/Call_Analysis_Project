```python
import pandas as pd
import numpy as np
datafile = "311_Service_Requests_2020.csv"
df = pd.read_csv(datafile)
df
```

```python
df10025 = df.loc[df['Incident Zip']==10025]
df10025 = df10025.dropna() #drop null values
```

```python
df['Complaint Type'].value_counts()
```

```python
# illegal parking complaints as a % of total complaints for full dataset
temp1 = len(df[df['Complaint Type'] == 'Illegal Parking'])/len(df)
temp1
```

```python
# illegal parking complaints as a % of total complaints for 10025 zip dataset
temp2 = len(df10025[df10025['Complaint Type'] == 'Illegal Parking'])/len(df10025)
temp2
```

```python
higher_parking_proportion = temp2 > temp1
higher_parking_proportion
```
